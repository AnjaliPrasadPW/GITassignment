```python
'''mY Tasks to Be Performed:


1 Based on what you have learnt in the class, do the following steps:

                 a. Create a new folder

                 b. Put the following files in the folder


                                    Code.txt

                                    Log.txt

                                    Output.txt


                 c. Stage the Code.txt and Output.txt files

                 d. Commit them

                 e. And finally push them to GitHub '''


 # a. Create a New Folder
     mkdir my-git-project
     cd my-git-project

 ## b. Create the Following Files
    touch Code.txt Log.txt Output.txt

 ### c. Stage Code.txt and Output.txt Files
      git init
     # now stage the files
     git add Code.txt Output.txt

 #### d. Commit the Files
      git commit -m "Added Code and Output files"


##### e. Push to GitHub
      #Connect your local folder to GitHub
      git remote add origin https://github.com/yourusername/my-git-project.

      #Push your commit to GitHub
      git push -u origin main



'''2. Please share the commands for the above point
  Tasks to Be Performed:

1. Create a Git working directory with feature1.txt and feature2.txt in the master branch

2. Create 3 branches develop, feature1 and feature2

3. In develop branch create develop.txt, do not stage or commit it

4. Stash this file and check out to feature1 branch

5. Create new.txt file in feature1 branch, stage and commit this file

6. Checkout to develop, unstash this file and commit

7. Please submit all the Git commands used to do the above step> '''


    #1. Create folder, start Git, add two files
         ## CREATE A FOLDER
         mkdir MyProject
         cd MyProject
         ## START GIT
         git init
         ##Added 2 files and saved them (committed)
         touch feature1.txt feature2.txt
         git add feature1.txt feature2.txt
         git commit -m "Added feature1.txt and feature2.txt in master branch"

     ## 2. Create 3 branches
     git branch develop
     git branch feature1
     git branch feature2

     ###3. Switch to develop branch and create a file (do not save)
     git checkout develop
     touch develop.txt

     #### 4. Stash the file and go to feature1 branch
     git stash
     git checkout feature1

    ####5. Create a file in feature1, stage it, and save (commit) it
    touch new.txt
    git add new.txt
    git commit -m "Added new.txt in feature1 branch"

    ##### 6. Go back to develop, get your old file back, and save it
    git checkout develop
    git stash apply
    git add develop.txt
    git commit -m "Committed develop.txt after unstashing"

    ######7. submit all the Git commands used to do the above step
    mkdir MyProject
    cd MyProject
    git init
    touch feature1.txt feature2.txt
    git add feature1.txt feature2.txt
    git commit -m "Added feature1.txt and feature2.txt in master branch"
    git branch develop
    git branch feature1
    git branch feature2
    git checkout develop
    touch develop.txt
    git stash
    git checkout feature1
    touch new.txt
    git add new.txt
    git commit -m "Added new.txt in feature1 branch"
    git checkout develop
    git stash apply
    git add develop.txt
    git commit -m "Committed develop.txt after unstashing"


 ''' Tasks to Be Performed:


1. Create a Git working directory, with the following branches:


               Develop

               F1

               f2


2. In the master branch, commit main.txt file

3. Put develop.txt in develop branch, f1.txt and f2.txt in f1 and f2 respectively

4. Push all these branches to GitHub

5. On local delete f2 branch

6. Delete the same branch on GitHub as well '''

#1: Create a Git working directory and branches
# Create a new folder and go inside it
mkdir git_project
cd git_project

# Initialize git repo
git init

# Create a file in master branch (default branch)
touch main.txt

# Add and commit main.txt in master
git add main.txt
git commit -m "Add main.txt in master branch"

# Create branches develop, F1, f2
git branch develop
git branch F1
git branch f2


##2: Add files in respective branches and commit
# Switch to develop branch
git checkout develop

# Create develop.txt
touch develop.txt

# Add and commit
git add develop.txt
git commit -m "Add develop.txt in develop branch"

# Switch to F1 branch
git checkout F1

# Create f1.txt
touch f1.txt

# Add and commit
git add f1.txt
git commit -m "Add f1.txt in F1 branch"

# Switch to f2 branch
git checkout f2

# Create f2.txt
touch f2.txt

# Add and commit
git add f2.txt
git commit -m "Add f2.txt in f2 branch"


#### 3: Push all branches to GitHub
# Add remote GitHub repo
git remote add origin <your-github-repo-URL>

# Push all branches
git push -u origin master
git push -u origin develop
git push -u origin F1
git push -u origin f2


### 4: Delete local f2 branch

git checkout master   # Make sure you are NOT on f2 branch
git branch -d f2      # Deletes local f2 branch


####5: Delete f2 branch on GitHub
git push origin --delete f2


''''4. Tasks to Be Performed:

1. Put master.txt on master branch, stage and commit

2. Create 3 branches: public 1, public 2 and private

3. Put public1.txt on public 1 branch, stage and commit

4. Merge public 1 on master branch

5. Merge public 2 on master branch

6. Edit master.txt on private branch, stage and commit

7. Now update branch public 1 and public 2 with new master code in private

8. Also update new master code on master

9. Finally update all the code on the private branch'''

# 1: Put master.txt on master branch, stage and commit
git init                            # Start a new Git repo
touch master.txt                    # Create the file
echo "This is master" > master.txt # Add content
git add master.txt                 # Stage the file (tell Git to track it)
git commit -m "Add master.txt"     # Save (commit) the file

##2: Create 3 branches: public1, public2, and private
git branch public1
git branch public2
git branch private


### 3: Put public1.txt on public1 branch, stage and commit
git checkout public1                  # Switch to public1 branch
touch public1.txt                     # Create the file
echo "This is public1" > public1.txt  # Add content
git add public1.txt
git commit -m "Add public1.txt"

###4: Merge public1 into master
git checkout master         # Switch back to master
git merge public1           # Merge public1 into master

####5: Merge public2 into master
git checkout master
git merge public2

#####6: Edit master.txt in private branch, stage and commit
git checkout private
echo "Private changes to master.txt" >> master.txt
git add master.txt
git commit -m "Update master.txt in private branch"

###### 7: Update public1 and public2 with new code from private
# Merge private into public1
git checkout public1
git merge private

# Merge private into public2
git checkout public2
git merge private

#######8: Update master with new code from private
git checkout master
git merge private

#########9: Update private with latest code from all branches
git checkout private
git merge master
git merge public1
git merge public2


'''5.Tasks to Be Performed:

1. Create a Git Flow workflow architecture on Git

2. Create all the required branches

3. starting from e feature branch, push the branch to the master, following the architecture

4. Push an urgent.txt on master using hotfix'''

#1: Create a Git Flow workflow architecture on Git
git init                      # Start a new Git project
git checkout -b master        # Create master branch
git checkout -b develop       # Create develop branch

## 2: Create all the required branches
git checkout -b feature/my-feature develop     # New feature branch from develop
git checkout -b release/v1.0 develop           # Create release branch
git checkout -b hotfix/urgent-fix master       # For fixing urgent bugs

###3: Starting from a feature branch, push the branch to the master, following the architecture
git checkout feature/my-feature
echo "This is my feature" > feature.txt
git add feature.txt
git commit -m "Add feature.txt"

git checkout develop
git merge feature/my-feature

git checkout release/v1.0
git merge develop

git checkout master
git merge release/v1.0




```
